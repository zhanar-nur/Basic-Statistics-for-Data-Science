The above output represents major statistical measures. It does exactly what it sounds like, describes the data. Describe returns a series of descriptive information. The Series describes following:

#### Count:
Count shows number of observations/rows for each variable.
#### Unique:
Unique for string columns represents the number of unique values, for example, for 'gender' variable there are three unique values in the dataset: male, female and other and that is why it's represented by 3 as per outcome of Unique column. This method doesn't list the number of unique values for numeric columns, since this is usually not particularly meaningful for numeric data.
#### Top:
Top gives the highest counted value of the categorical variables. For example, Top for 'gender' variable is represented by Female and in order to check it, run df['gender'].value_counts() which will give the following output: Female 2994, Male 2115 and Other 1.
#### Freq:
Freq shows the frequency of top values, as with the example described above, the frequency of top value for Female in 'gender' column is 2,994.
#### Mean:
Mean is a measure of the central tendency of a probability distribution. Specifically, it's the sum of the values divided by the number of values. Mean can be used with both discrete and continuous variables, although its use is most often with continuous variables. It's not recommended to use mean to measure central tendency if there are outliers in a dataset, as it is particularly susceptible to the influence of outliers. If there are outliers in a dataset, the mean value might not be the best way to accurately reflect the measure of the central tendency.
#### Std:
Std stands for standard deviation. The standard deviation shows how the values are distributed relative to the mean. The closer the standard deviation is to 0, the more reliable the mean. Moreover, a standard deviation close to 0 indicates little variability in the data. In our example, 'bmi' variable has the lowest variability.
#### Min:
Min or least element of a dataset.
#### 25%:
25% percentile, or also known as the first, or lower quartile. It shows the value at which 25% of given observations fall. Percentiles are used to understand and interpret data. For example, a 22-year-old observed from our data is in the 25th percentile of his or her age. This means that 25 percent of all observed are under 22 years old.
#### 50%:
50% percentile, or also known as the Median. It shows the value at which 50% of given observations fall. In other words, the median divides the data set in half.  Half of the values lie below the median and another half lie above the median. The median is less affected by outliers in comparison to mean.
#### 75%:
75% percentiles, or also known as the third, or upper quartile. It shows the value at which 25% of the observations lie above that value and 75% of the observations lie below that value. 
#### Max:
Maximum element of a dataset.
### Other:
#### Mode:
Mode - the name of this term speaks for itself, as it's the value that occurs most in the sample. Normally, the mode is used for categorical variables in order to find out which is the most common category.

```python
# histogram for 'avg_glucose_level'

df_avg_glucose_level = df['avg_glucose_level']
df_avg_glucose_level.plot.hist(alpha=0.5, bins=20, grid=True, legend=None, density=True, color='blue', edgecolor='black')
mean = df_avg_glucose_level.mean()
median = df_avg_glucose_level.median()
mode = df_avg_glucose_level.mode()
variance = df_avg_glucose_level.var()
sigma = math.sqrt(variance)
min = df_avg_glucose_level.min()
max = df_avg_glucose_level.max()
x = np.linspace(mean -3*sigma, mean + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mean, sigma),color = 'r')
plt.title('Glucose level in blood', fontdict = {'fontweight':'bold', 'fontsize': 16})
plt.xlabel('Glucose level in blood')
plt.ylabel('Frequency')
plt.show()
print('Mean = ', mean)
print('Median = ', median)
print('Mode = ', mode)
print('SD = ', sigma)
print('Min = ', min)
print('Max = ', max)
```

The above histogram visualizes the distribution of 'glucose level in blood'. The taller the bar, the more data falls into that range of the histogram. In our case, the highest number of observations falls at 93.33 which is the mode. As we can see from the statistics of the histogram, the standard deviation is 45.3 meaning that the distribution of the input data relative to the mean has high variability as "average" distance or each individual measurement is from the average equals to 45.3. The above histogram is not symmetrical, it is skewed to the right (positively skewed distribution). In this case, the median is generally considered to be the best representative of the central tendency of the dataset.

```python
# statistical test: T-test for comparison of 'bmi' by 'gender'

from scipy.stats import ttest_ind
df_male = df.loc[df['gender'] == 'Male']
male_bmi = df_male['bmi']
df_female = df.loc[df['gender'] == 'Female']
female_bmi = df_female['bmi']
ttest,pval = ttest_ind (male_bmi, female_bmi, equal_var = False)
print('t-test statistic =', ttest, 'p-value=', pval)
print('male bmi mean')
print(df_male['bmi'].mean())
print('female bmi mean')
print(df_female['bmi'].mean())
```

The above output of t-test compares the means of two groups (t-test is for numeric variables) in order to tell how significant the differences between groups are. t-test statistics is the measure that shows the difference between two groups. The score of t-test is at -1.92 means that the groups are 1.92 times as different from each other as they are within each other. The smaller the t-score, the more similarity there is between groups. A p-value of the above output indicates the probability that the results of the dataset occurred by chance. Since p-value is 0.054782501247195234 or 5%, it's considered low p-value meaning that there is only a 5% probability that the results from the test happened by chance and the data is valid. 

```python
# statistical test: Chi-test for 'gender' by 'smoking_status'

from scipy.stats import chisquare
crosstab_gender_smoke = pd.crosstab(df['gender'], df['smoking_status'])
gender_smoke_chi2 = stats.chi2_contingency(crosstab_gender_smoke, correction = False)
print('observed counts')
print(crosstab_gender_smoke)
print('expected counts')
print(gender_smoke_chi2[3])
print('Chi-Square')
print(gender_smoke_chi2[0])
print('p-value')
print(gender_smoke_chi2[1])
```

The above Chi2-test is used to determine whether there is a correlation between 'gender' and 'smoking_status'. The test is used for categorical variables. Another use of the test is to find out the difference between the observed and the expected data. 
The p-value is 15.67% which means that the null hypothesis is not rejected at 95% level of confidence. It means that 'gender' and 'smoking_status' are independent. 

```python
# ANOVA for 'smoking_status' & 'avg_glucose_level'
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import MultiComparison
from statsmodels.stats.multicomp import pairwise_tukeyhsd
ANOVA_Smoke_Glucose = ols('avg_glucose_level ~ smoking_status', data=df).fit()
# ANOVA 'smoking_status' & 'avg_glucose_level' summary
SimpleANOVA_smoke_Glucose = sm.stats.anova_lm(ANOVA_Smoke_Glucose, type = 2)
print(SimpleANOVA_smoke_Glucose)
MC_Smoke_Glucose = MultiComparison(df['avg_glucose_level'], df['smoking_status'])
Tukey_Smoke_Glucose = MC_Smoke_Glucose.tukeyhsd()
print(Tukey_Smoke_Glucose)
print(MC_Smoke_Glucose.groupsunique)
```

The above analysis of variance test, or ANOVA is used as a generalization for more than 2 groups to compare the means between more than two groups. The second part of the test (post-hoc test) is used to find out where the difference among groups as ANOVA tdoes not tell where the difference is, it only shows if there is some difference among groups. 
p-value of ANOVA is greater than 0.05 meaning that the null hypothesis is accepted, so there is no significant difference among the groups. 
Post-hoc test shows if there is a statistically significant difference among the groups. (However, our example is not good demonstration for ANOVA. I will have the separate example for ANOVA). The basic meaning of the second output is checking p-values of two groups. As we can see from the above output, there are two combinations of groups that have significant difference (0.1216 and 0.9 respectively).

```python
# checking for statistical assumptions - Normality

from scipy.stats import shapiro
stat, p = shapiro(df_male['avg_glucose_level'])
print('Shapiro Statistic=%.3f, p=%.3f' % (stat, p))
stat, p = shapiro(df_female['avg_glucose_level'])
print('Shapiro Statistic=%.3f, p=%.3f' % (stat, p))
```

The above code is used in order to identify the Normality of the distribution that is usually checked before ANOVA test. One of the method to check it is Shapiro-Wilk test. P-value is at 0.000, meaning that 'avg_glucose_level' data is not normally distributed.

```python
# histogram for 'bmi' of males
male_bmi.plot.hist(alpha=0.5, bins = 50, grid=True, legend=None, density=True, color='blue', edgecolor='black')
mean = male_bmi.mean()
variance = male_bmi.var()
sigma = math.sqrt(variance)
x = np.linspace(mean - 3*sigma, mean + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mean, sigma),color = 'r')
plt.title('BMI by Genders', fontdict = {'fontweight':'bold', 'fontsize': 16})
plt.xlabel('BMI for Males')
plt.ylabel('Frequency')
plt.show()
```

```python
# qq plot to check normality
from statsmodels.graphics.gofplots import qqplot
qqplot(male_bmi, line = 'q')
plt.title('BMI by Genders', fontdict = {'fontweight':'bold', 'fontsize': 16})
plt.show()
```

```python
# qq plot to check normality
from statsmodels.graphics.gofplots import qqplot
qqplot(female_bmi, line = 'q')
plt.title('BMI by Genders', fontdict = {'fontweight':'bold', 'fontsize': 16})
plt.show()
```

The above QQ plots are used to demonstrate the normality of the distribution as well. The deviations from the straight line are minimal. This indicates normal distribution.

```python
# homogeneity of variances
stat, p = stats.levene(male_bmi, female_bmi)
print('Levene Statistic = %.3f, p=%.3f' % (stat, p))
```

ANOVA test requires another assumption for homogeneity. Levene Statistic is one of the methods to define it. This test shows whether the two means for male 'bmi' and female 'bmi' have equal variances. In the above output, the p-value is 0.000, and that means that equal variance is not assumed.

```python
# normality distributed residuals for 'smoking_status' & 'avg_glucose_level'
ANOVA_Smoke_Glucose = ols('avg_glucose_level ~ smoking_status', data=df).fit()
ANOVA_Smoke_Glucose.summary()
res = ANOVA_Smoke_Glucose.resid
stat, p = shapiro(res)
print('Shapiro Statistic=%.3f, p=%.3f' % (stat, p))
fig = sm.qqplot(res, line = 'q')
plt.show
```

The above QQ plot is used to demonstrate the normality of the distribution for 'avg_glucose_level' and 'smoking_status'. The deviations from the straight line are significant. This indicates no normal distribution present.
